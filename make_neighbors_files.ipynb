```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm
```

```python
books_quantile = 0.95
clients_quantile = 0.5
```

```python
books_sim = pd.read_csv('books_sim_2000.csv')
books_sim['Unnamed: 0'] = books_sim['Unnamed: 0'].astype('str')
books_sim = books_sim.set_index('Unnamed: 0')
books_sim.head()
```

```python
books_threshold = pd.Series(books_sim[(books_sim!=1)&(books_sim!=0)].values.flatten()).quantile(books_quantile)
books_threshold
```

```python
books_neighbors = pd.DataFrame({'id':books_sim.index})
neighbors = []
for ind in tqdm(books_neighbors['id'].values):
    n_list = books_sim.columns[books_sim.loc[ind]>books_threshold].values
    n_list = n_list[n_list != ind]
    try:
        neighbors.append((n_list + ',').sum()[:-1])
    except:
        neighbors.append('')
books_neighbors['neighbors'] = neighbors
books_neighbors = books_neighbors.set_index('id')
```

```python
print(books_neighbors.shape)
books_neighbors.head()
```

```python
clients_sim = pd.read_csv('clients_sim_5000.csv')
clients_sim['userid'] = clients_sim['userid'].astype('str')
clients_sim = clients_sim.set_index('userid')
print(clients_sim.shape)
clients_sim.head()
```

```python
clients_threshold = pd.Series(clients_sim[(clients_sim!=1)&(clients_sim!=0)].values.flatten()).quantile(clients_quantile)
clients_threshold
```

```python
clients_neighbors = pd.DataFrame({'id':clients_sim.index})
neighbors = []
for ind in tqdm(clients_neighbors['id'].values):
    n_list = clients_sim.columns[clients_sim.loc[ind]>clients_threshold].values
    n_list = n_list[n_list != ind]
    try:
        neighbors.append((n_list + ',').sum()[:-1])
    except:
        neighbors.append('')
clients_neighbors['neighbors'] = neighbors
clients_neighbors = clients_neighbors.set_index('id')
```

```python
print(clients_neighbors.shape)
clients_neighbors.head()
```

```python
books_neighbors.to_csv('books_neighbors.csv')
clients_neighbors.to_csv('clients_neighbors.csv')
```
